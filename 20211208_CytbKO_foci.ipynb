```python
import pandas as pd
import seaborn as sns
```

```python
df_wt = pd.read_csv('Results_20211208_1121wtSlide1.csv', names=['Cell_ID','start_z','stop_z','prominence','foci_count','label'])
df_wt['label']='wt'
df_ko = pd.read_csv('Results_20211208_1121koSlide1.csv', names=['Cell_ID','start_z','stop_z','prominence','foci_count','label'])
df_ko['label']='ko'
df = df_wt.append(df_ko)
```

```python
# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=df[df['prominence']>100], x="prominence", y="foci_count", hue="label",
               split=True, inner="quart", linewidth=1)
sns.despine(left=True)
```

```python
dff = df[(df['prominence']<200000) & (df['prominence']>10000)]
```

```python
# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=dff, x="label", y="foci_count", hue="prominence",
               split=False, inner="quart", linewidth=1)
sns.despine(left=True)
```

```python
# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="prominence", y="foci_count",
            hue="label", palette=["m", "g"],
            data=df[df['prominence']<200000])
sns.despine(offset=10, trim=True)
```

```python
# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="label", y="foci_count",
            hue="prominence",
            data=df)
sns.despine(offset=10, trim=True)
```

```python
DATA = df[df['label']=='ko']

# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="prominence", y="foci_count",
            hue="label",
            data=DATA)
#sns.despine(offset=10, trim=True)

# Add in points to show each observation
sns.stripplot(x="prominence", y="foci_count", hue="label", data=DATA,
              size=4, linewidth=0)
```
